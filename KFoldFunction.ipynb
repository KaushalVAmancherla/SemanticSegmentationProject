```python
from sklearn.model_selection import KFold
import numpy as np
```

```python
def kFoldSplit(img_list,fold_num):
  list_dict = []

  kfold = KFold(5,False,None)
  data = np.array(img_list)

  for train, test in kfold.split(data):
	#print(len(data[train]), len(data[test]))
    train_set = data[train]

    idx = len(data[test])//2

    val_set = data[test][:idx]
    test_set = data[test][idx:]

    dict_split = {"train":train_set,"validation":val_set,"test":test_set}
    list_dict.append(dict_split)

  return list_dict[fold_num]
```
